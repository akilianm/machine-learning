### Mutual Information (categorical column importance) - tells you how important each variable is. Tells us how much we can learn about one variable if we know the value of another

```python
from sklearn.metrics import mutual_info_score
```

```python
mutual_info_score(df_full_train.churn, df_full_train.contract)
```

```python
mutual_info_score(df_full_train.churn, df_full_train.gender)
```

```python
mutual_info_score(df_full_train.churn, df_full_train.partner)
```

```python
# Create function for mutual_info since apply only takes one argument

def mutual_info_churcn_score(series):
    return mutual_info_score(series, df_full_train.churn)

# Use apply function

mi= df_full_train[categorical].apply(mutual_info_churcn_score) #iterates through every column in categorical and applies mutual_info_churn_score function to it
mi.sort_values(ascending=False)
```
